```python
import pandas as pd
df = pd.read_csv('./2019_2020_top_10.csv', index_col='pusher_id')
df.head()
```

```python
top10 = list(set(df.index)) # top 8 actually
', '.join(top10)
```

```python
# 斷詞 function
from ckiptagger import WS

def segmentation(sentence_list):
    
    # 使用 CPU：
    ws = WS("./data")
    
    word_sentence_list = ws(sentence_list)
    
    # deallocate memory
    del ws
    
    return word_sentence_list
```

```python
# 取出某人所有的留言
sentences = list(map(lambda x: x[0], df.loc[top10[1]][['push_content']].values.tolist()))

# 送進斷詞 func
word_sentence_list = segmentation(sentences)
word_sentence_list = list(map(lambda x: x[0], word_sentence_list))

word_sentence_list
```

```python
occurence = [[uniqueWord, word_sentence_list.count(uniqueWord)] for uniqueWord in list(set(word_sentence_list))]
sorted(occurence, key = lambda x: x[1], reverse=True)
```
