```python
import pandas as pd
import itertools
df = pd.read_csv('./2019_2020_top_10.csv', index_col='pusher_id')
df.head()
```

```python
top10 = list(set(df.index)) # top 8 actually
', '.join(top10)
```

```python
# 斷詞 function
from ckiptagger import WS, POS

def segmentation(sentence_list):
    
    # 使用 CPU：
    ws = WS("./data")
    pos = POS("./data")

    word_sentence_list = ws(sentence_list)
    pos_sentence_list = pos(word_sentence_list)
    
    # deallocate memory
    del ws
    del pos
    
    return word_sentence_list, pos_sentence_list
```

```python
# 取出某人所有的留言
sentences = list(map(lambda x: x[0], df.loc[top10[0]][['push_content']].values.tolist()))
# 送進斷詞 func
word_sentence_list, pos_sentence_list = segmentation(sentences)
# flatten
word_sentence_list = list(itertools.chain(*word_sentence_list))
pos_sentence_list = list(itertools.chain(*pos_sentence_list))
```

```python
# 合併單詞list和詞性list
word_pos_sentence_list = list(zip(word_sentence_list, pos_sentence_list))
# 去除重複並計算數量
occurence = [[uniqueWord, word_pos_sentence_list.count(uniqueWord)] for uniqueWord in list(set(word_pos_sentence_list))]
# 由大到小排序
occurence.sort(key = lambda x: x[1], reverse=True)
occurence
```
